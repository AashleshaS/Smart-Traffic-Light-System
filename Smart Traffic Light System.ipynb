```python
# STEP 3: Setup Kaggle credentials
!mkdir -p ~/.kaggle
!cp kaggle.json ~/.kaggle/
!chmod 600 ~/.kaggle/kaggle.json

# STEP 4: Download Highway Traffic Videos dataset
!kaggle datasets download -d aryashah2k/highway-traffic-videos-dataset
!unzip -qo highway-traffic-videos-dataset.zip -d traffic_videos

# The rest of your vehicle detection code goes here
import cv2
import time
from ultralytics import YOLO

# Load YOLOv8n model with verbose=False
model = YOLO("yolov8n.pt", verbose=False)

video_path = "traffic_videos/video/cctv052x2004080516x01638.avi"
cap = cv2.VideoCapture(video_path)

if not cap.isOpened():
    print(f"Error: Could not open video file {video_path}")
else:
    frame_count = 0
    low_density_threshold = 5
    high_density_threshold = 20

    low_light_time = 15
    medium_light_time = 30
    high_light_time = 60

    print("--- Starting Traffic Light Simulation ---")

    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            break

        results = model(frame, verbose=False)
        vehicle_count = 0
        for result in results:
            boxes = result.boxes
            vehicle_count += sum(
                1 for box in boxes
                if model.names[int(box.cls[0])] in ["car", "truck", "bus", "motorbike"]
            )

        traffic_density = "Low"
        green_light_duration = low_light_time

        if vehicle_count >= high_density_threshold:
            traffic_density = "High"
            green_light_duration = high_light_time
        elif vehicle_count >= low_density_threshold:
            traffic_density = "Medium"
            green_light_duration = medium_light_time
        else:
            traffic_density = "Low"
            green_light_duration = low_light_time

        print(f"Frame {frame_count}: Vehicle count = {vehicle_count}, Density = {traffic_density}, Green Light Duration = {green_light_duration}s")

        frame_count += 1
        time.sleep(1)

    cap.release()
    print("--- Simulation Finished ---")
```
